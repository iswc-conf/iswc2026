```python
def text_to_li(text):
    lines = text.splitlines()
    return "\n".join(f"<li>{line}</li>" for line in lines if line.strip())


# Example usage
input_text = """
Registration fees can be paid by credit card.
Accepted cards: VISA, MasterCard.
Payment is confirmed at the time of registration.

"""

output = text_to_li(input_text)
print(output)
```
